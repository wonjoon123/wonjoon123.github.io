# 알고리즘 함수 작성

이곳에 알고리즘 코드를 작성해주시면 됩니다. 알고리즘 코드 작성 및 file_path 이외의 부분의 코드에는 변형을 가하시면 안됩니다.

```python
def your_algorithm(water_list):
  
    def trap_water(water_list):
        n = len(water_list)
        if n == 0:
            return 0
        
        left_max = [0] * n
        right_max = [0] * n
        
        left_max[0] = water_list[0]
        for i in range(1, n):
            left_max[i] = max(left_max[i - 1], water_list[i])
        
        right_max[n - 1] = water_list[n - 1]
        for i in range(n - 2, -1, -1):
            right_max[i] = max(right_max[i + 1], water_list[i])
        
        trapped_water = 0
        for i in range(n):
            water_level = min(left_max[i], right_max[i])
            if water_level > water_list[i]:
                trapped_water += water_level - water_list[i]
        
        return trapped_water

  # 0을 기준으로 구간을 나누어 저장

    segments = []
    current_segment = []
    
    for height in water_list:
        if height == 0:
            if current_segment:
                segments.append(current_segment)
                current_segment = []
        else:
            current_segment.append(height)
    
    if current_segment:
        segments.append(current_segment)

  # 각 구간에서 물의 양을 계산하고 합산
    water_storage = 0
    for segment in segments:
        water_storage += trap_water(segment)
    
    
    return water_storage
```

#채점 코드

input에 대한 본인 알고리즘의 답과 실제 정답을 대조하는 코드

아래는 총 1000개의 input에 대한 정답과 오답 여부를 파악할 수 있는 코드

실제 채점에서는 더 많은 input으로 점수를 채점합니다.

```python
score = 0

for i in tqdm(range(len(test_df))):
  list_1 = test_df['input'][i]
  list_1 = ast.literal_eval(list_1)
  answer = test_df['answer'][i]

  your_answer = your_algorithm(list_1)



  if answer==your_answer:
    score = score + 1
  else :

    pass

print('1000개 중 맞힌 개수 : ' , score)
```

#알고리즘의 효율성을 파악하기 위한 코드

```python
#길이가 1000만이고 0~4사이의 정수 값을 갖는 무작위 리스트 생성
long_list = [random.randint(0,4) for _ in range(10000000)]
```

```python
start_time = time.time()

your_algorithm(long_list)

end_time = time.time()

print('처리에 걸린 시간 : ', end_time - start_time)
```
